# Import Required Libraries
Import necessary libraries for creating the humanoid maze-medium environment and plotting.

```python
import matplotlib.pyplot as plt
import numpy as np
import sys
sys.path.append('/home/jennifer/aorl')
from wrappers.datafuncs_utils import make_env_and_datasets
```

# Create the Humanoid Maze-Medium Environment
Initialize the humanoid maze-medium environment for task3 and task4.

```python
# Create environments for task3 and task4
env_task3, _, _ = make_env_and_datasets('humanoidmaze-medium-navigate-singletask-task3-v0', dataset_path=None, use_oracle_reps=True)
env_task4, _, _ = make_env_and_datasets('humanoidmaze-medium-navigate-singletask-task4-v0', dataset_path=None, use_oracle_reps=True)

print("Task3 info:", env_task3.task_infos)
print("Task4 info:", env_task4.task_infos)
```

# Extract Maze Cells and Task Positions
Extract the maze layout and task positions.

```python
# Assuming the maze is a grid, extract positions
# For simplicity, let's assume oracle reps are 2D positions
# We need to get all possible positions or the maze walls

# Reset env to get initial positions
obs_task3 = env_task3.reset()
goal_task3 = env_task3.task_infos['goal']
start_task3 = obs_task3[:2]  # Assuming first two are position

obs_task4 = env_task4.reset()
goal_task4 = env_task4.task_infos['goal']
start_task4 = obs_task4[:2]

print("Task3 start:", start_task3, "end:", goal_task3)
print("Task4 start:", start_task4, "end:", goal_task4)

# To plot all cells, we might need to get the maze layout
# For now, let's assume a grid and plot positions
```

# Plot the Maze with Marked Positions
Plot the maze cells and mark the task positions.

```python
# Plot
fig, ax = plt.subplots()

# Plot start and end for task3
ax.scatter(start_task3[0], start_task3[1], color='blue', label='Task3 Start', s=100, marker='o')
ax.scatter(goal_task3[0], goal_task3[1], color='blue', label='Task3 End', s=100, marker='x')

# Plot start and end for task4
ax.scatter(start_task4[0], start_task4[1], color='red', label='Task4 Start', s=100, marker='o')
ax.scatter(goal_task4[0], goal_task4[1], color='red', label='Task4 End', s=100, marker='x')

ax.legend()
ax.set_title('Humanoid Maze-Medium: Task3 and Task4 Positions')
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.show()
```
